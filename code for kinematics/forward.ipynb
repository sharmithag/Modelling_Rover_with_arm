```python
'''FORWARD KINEMATICS AND ITS VALIDATION'''

from sympy import *
from sympy.physics.mechanics import dynamicsymbols
from math import pi


#Function for calculating the final transformation matrix
def trans_matrix (th1,th2,th3,r,d1,d2):
    T01 = Matrix([[cos(th1), 0, -sin(th1), 0], [sin(th1), 0, cos(th1), 0], [0, -1, 0, d1], [0, 0, 0, 1]])
    T12 = Matrix([[cos(th2), -sin(th2), 0,r*cos(th2)], [sin(th2),cos(th2), 0,r*sin(th2)], [0, 0, 1, 0], [0, 0, 0, 1]])
    T23 = Matrix([[cos(th3), -sin(th3), 0,0], [sin(th3),cos(th3), 0,0], [0, 0, 1, d2], [0, 0, 0, 1]])
   
    final_matrix = (T01 * T12 * T23)
    pprint(final_matrix)
th1,th2,th3,r,d1,d2=dynamicsymbols('th1 th2 th3 r d1 d2')
trans_matrix (th1,th2,th3,r,d1,d2)
ths = [0, 0, 0]
print(ths)

for i in range(0, 3):
    print("\n Configuration #", i+1," when th", i, " is pi/2\n")
    if i == 0:
        ths[i] = pi/2
    else:
        ths[i] = pi/2
        ths[i-1] = 0
    trans_matrix (ths[0], ths[1], ths[2],-788.40,130,250)
    
```
